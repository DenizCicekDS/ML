```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
fig, ax = plt.subplots()
ax.plot(age, salary)
plt.show()
```

```python
plt.plot(age, salary, label="1")
plt.plot(age, salary_2, label = "2")
plt.legend()
```

```python
def multi_graf(rows,cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(rows):
        for j in range(cols):
            ax[i,j].plot(age, salary)
            plt.tight_layout()
multi_graf(3,2)
```

```python
fig, ax = plt.subplots()
langs = ['C', 'Python', 'PHP', "HTML&Css", "sql", "git&linux"]
students = [3,25,1,14,22,18]
ax.bar(langs,students)
plt.show()
```

```python
tips.groupby('day').sum()
```

```python
tips.groupby('day')['total_bill'].sum()
```

```python
day = tips.day.unique()
total_bill= tips.groupby('day')['total_bill'].sum()

fig, ax = plt.subplots()

ax.bar(day, total_bill)
plt.show()
```

```python
day = tips.groupby("day")["total_bill"].sum().index
total_bill = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day,total_bill)
plt.show()
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(day,tip)
plt.show()
```

```python
sex = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(sex,tip)
plt.show()
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day,total, label="total")
ax.bar(day,tip, label="tip")
plt.legend()
plt.show()
```
